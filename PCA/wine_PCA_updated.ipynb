```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 200
%matplotlib inline
```

```python
data = pd.read_csv('wine.csv')
data
```

```python
df = data.iloc[:,1:]
df
```

```python
plt.scatter(df['Alcohol'],df['Alcalinity'])
plt.xlabel('Alcohol')
plt.ylabel('Alcalinity')
```

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(df)
x
```

```python
pca = PCA()
```

```python
pcs = pca.fit_transform(x)
pcs
```

```python
pd.DataFrame(pcs)
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
data = pd.DataFrame(pcs)
data
```

```python
df = data.iloc[:,:3]
df
```

```python
df.rename(columns={0 : 'PC1', 1 : 'PC2', 2 : 'PC3'}, inplace=True)
df
```

```python
from scipy.cluster import hierarchy
```

```python
lk = hierarchy.linkage(df, method='complete')
```

```python
dendrogram = hierarchy.dendrogram(lk)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=3)
```

```python
ypred = hc.fit_predict(df)
ypred
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df,ypred)
```

```python
plt.scatter(df['PC1'],df['PC2'], c=ypred, cmap='rainbow')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
#Hierarchical clustering gives three clusters as optimum number of clusters
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python
km = KMeans(n_clusters=3)
```

```python
ypred = km.fit_predict(df)
ypred
```

```python
silhouette_score(df,ypred)
```

```python
plt.scatter(df['PC1'],df['PC2'], c=ypred, cmap='rainbow')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
#KMeans clustering also gives three clusters as optimum number of clusters. Hence, we have received same number of clusters.
```
