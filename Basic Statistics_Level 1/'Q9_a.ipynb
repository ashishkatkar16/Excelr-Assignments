```python
import pandas as pd
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Q9_a.csv', index_col=0)
```

```python
df
```

```python
skew_speed = skew(df['speed'])
kurtosis_speed = kurtosis(df['speed'])
print("Skewness of speed:", skew_speed)
print("Kurtosis of speed:", kurtosis_speed)
skew_dist = skew(df['dist'])
kurtosis_dist = kurtosis(df['dist'])
print("Skewness dist:", skew_dist)
print("Kurtosis dist:", kurtosis_dist)
```

```python
plt.hist(df['speed'], color='skyblue', edgecolor='black')
```
