```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
#Problem 1(Forestfires)
```

```python
df = pd.read_csv('forestfires.csv')
df
```

```python
df.drop(columns=['month','day'], inplace=True)
df
```

```python
df.replace(['small','large'],[0,1], inplace=True)
df
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated(keep=False)]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df
```

```python
x= df.iloc[:,:-1]
y = df['size_category']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
xtrain
```

```python
ytrain
```

```python
ann = Sequential()

ann.add(Dense(units = 10, activation = 'relu'))
ann.add(Dense(units = 1, activation = 'sigmoid'))

ann.compile(optimizer = 'adadelta', loss = 'binary_crossentropy', metrics ='accuracy')

ann.fit(xtrain,ytrain,epochs = 100)
```

```python
ypred = ann.predict(xtest)
ypred = ypred>0.5
ypred
```

```python
print(classification_report(ytest,ypred))
```

```python
import keras_tuner as kt
```

```python
def hyper(hp):
    model = Sequential()
    counter = 0
    for i in range(hp.Int('num_layers', min_value=1, max_value=10)):
        if counter == 0:
            model.add(Dense(
                hp.Int('units' + str(i), min_value=8, max_value=128, step=8),
                activation=hp.Choice('activation' + str(i), values=['relu', 'tanh', 'sigmoid']),
                input_dim=28
            ))
        else:
            model.add(Dense(
                hp.Int('units' + str(i), min_value=8, max_value=128, step=8),
                activation=hp.Choice('activation' + str(i), values=['relu', 'tanh', 'sigmoid'])
            ))
        counter += 1
    
    model.add(Dense(units=1, activation='sigmoid'))
    model.compile(
        optimizer=hp.Choice('optimizer', values=['adam', 'rmsprop', 'sgd']),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
tuner = kt.RandomSearch(hyper,
                        objective = 'val_accuracy',
                        max_trials = 3)
```

```python
tuner.search(xtrain,ytrain,epochs = 5, validation_data = (xtest,ytest))
```

```python
tuner.get_best_hyperparameters()[0].values
```

```python
model = tuner.get_best_models(num_models=1)[0]
```

```python
model.fit(xtrain,ytrain, epochs=100, validation_data=(xtest,ytest))
```

```python
ypred = model.predict(xtest)
ypred = ypred>0.5
ypred
```

```python
print(classification_report(ytest,ypred))
```

```python
x = df.drop(columns='TEY')
y = df['TEY']
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
ann = Sequential()

ann.add(Dense(units=30, activation='relu'))
ann.add(Dense(units=20, activation='relu'))

ann.add(Dense(units=1))

ann.compile(optimizer='adam',loss = 'mse')

ann.fit(xtrain,ytrain, epochs = 100, validation_data = (xtest,ytest))
```

```python
ypred = ann.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(ytest,ypred)
```
